```python
import numpy as np
import pandas as pd
```

```python
# importing data from *.csv to datafram 
data= pd.read_csv('icecreamratings.csv')
```

```python
# varifying if data is in data veriable
data.head() # head function will display first 5 recods from dataset
```

```python
# set the index in dataset
data=data.set_index('Date')
data
```

```python
# very simple plot
data.plot() # it will go with default settings 
```

```python
# override the kind parameter
data.plot( kind='line') # it will display a line graph, also. 
```

```python
# override the parameters
data.plot( kind='line',subplots=True) # it will display a line graph, and subplots. 
```

```python
# override the parameters
data.plot( kind='line', title='Ice Cream Ratings') # adding plot kind and title. 
```

```python
# override the kind parameter
data.plot( kind='line',xlabel='Daily Ratings',ylabel='Scores') #updating xlabel and ylabel . 
```

```python
# # override the parameters
data.plot( kind='bar', title='Ice Cream Ratings') # adding plot kind as barplot and title. 
```

```python
# turning barplot into stacke form
# override the parameters
data.plot( kind='bar',stacked=True, title='Ice Cream Ratings') # adding plot kind as barplot, stacked and title. 
```

```python
# specifying a diamension to plot
# override the parameters
data['Flavor Rating'].plot( kind='bar', title='Ice Cream Ratings') # adding  Flavor Rating	 as a  diamension and title. 
```

```python
# visualizing a horizontal bar chart
# override the parameters
data.plot.barh( stacked=True, title='Ice Cream Ratings') # stacked horizontal bar chart with  title. 
```

```python
# Scatter plot
data.plot.scatter(x='Texture Rating',y='Flavor Rating',s=150 , c='blue') # addding size and color 
```

```python
# Scatter plot
data.plot.hist() # with default parameters 
```

```python
# Scatter plot
data.plot.hist(bins=20) # overriding bins=20
```

```python
# Scatter plot
data.boxplot() # with default parameters 
```

```python
# Scatter plot
data.boxplot('Flavor Rating') #only for one diamension 
```

```python
# area  plot
data.plot.area() # with default parameters 
```

```python
# area  plot
data.plot.area(figsize=(10,5)) # adding figsize 
```

```python
# pie plot
data.plot.pie(y='Overall Rating',figsize=(8,7)) # with default parameters 
```
